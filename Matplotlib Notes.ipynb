# Figures

Matplotlib's top-level plotting object is a **figure**. 

Within a figure there are one or more **axes**. When you plot something, such as a bar chart, the plotting is done on an axes within a figure.

NOTE: When we say axes, we don't mean x and y axis.  An entire graph is an axes.  So one figure can have three different graphs, each being an axes.  EG axes1 being a bar plot, axes2 being a scatter plot, axes3 being a line plot.

A figure can have a **figure title**. EG "A figure with three axes". Each axes can also have its own **axes title**. EG: "An axes with bar plot", "An axes with scatter plot and grid", and "An axes with line plot". 

Each axes has an x-axis and a y-axis.

An axis can have **major ticks and minor ticks**. Each tick can have a tick label.

An axes can have a **grid**, which is a series of horizontal and vertical lines in the plotting area. 

An axes can have a **legend**, and the legend can have a **legend title and legend labels**. EG you can have a legend title like "Legend", and whose labels are "Line 1" and "Line 2".

There are many different types of plots. Some examples:

* Line
* Area
* Pie
* Bar
* Histogram
* Box
* Density
* Scatter
* Hexbin

# plylot direct

```python
import matplotlib.pyplot as plt

data = {
    'NSW': {'Q1': 3.2, 'Q2': 3.4, 'Q3': 3.4, 'Q4': 3.6},
    'VIC': {'Q1': 3.5, 'Q2': 3.4, 'Q3': 3.0, 'Q4': 3.1},
}

# Create a new figure and call it 'fig'
fig = plt.figure()

# Add an axes to the figure and call it 'ax'
ax = fig.add_subplot()

# Add a LINE plot to ax
#     The .PLOT after ax specifies that this is a line graph if we leave it blank and thus get the default
#     ax.bar would give us a bar graph, ax.scatter a scatter graph etc
# Use the quarters as the x-values and the NSW percentages as the y-values
ax.plot(data['NSW'].keys(), data['NSW'].values())

# Add a line plot to ax
# This time use the VIC percentages as the y-values
ax.plot(data['VIC'].keys(), data['VIC'].values())

# Optional command to save the plot as .png file, necessary in Ed
#fig.savefig('plot.png')
```

## Adding some features

```python
fig = plt.figure()

# Add a figure title
fig.suptitle('Unemployment Rates')

ax = fig.add_subplot()

# Specify labels this time (label = )
ax.plot(data['NSW'].keys(), data['NSW'].values(), label='NSW')
ax.plot(data['VIC'].keys(), data['VIC'].values(), label='VIC')

# Show a legend
ax.legend()

# Specify axis labels 
ax.set_xlabel('Quarter')
ax.set_ylabel('Unemployment (%)')

#fig.savefig('plot.png')
```

## Using Multiple Axes

```python
fig = plt.figure()
fig.suptitle('Unemployment Rates')

# Add an axes. It's the first axes of a 1 x 2 grid of axes.
ax1 = fig.add_subplot(1, 2, 1) # Number of rows, Number of columns, Position of this axes

# No need for a label this time
ax1.plot(data['NSW'].keys(), data['NSW'].values())

# Specify a title for the axes, and labels for the x- and y-axis
ax1.set_title('NSW')
ax1.set_xlabel('Quarter')
ax1.set_ylabel('Unemployment (%)')

# Add an axes. It's the second axes of a 1 x 2 grid of axes.
ax2 = fig.add_subplot(1, 2, 2) # Number of rows, Number of columns, Position of this axes

# No need for a label this time
ax2.plot(data['VIC'].keys(), data['VIC'].values())

# Specify a title for the axes, and labels for the x- and y-axis
ax2.set_title('VIC')
ax2.set_xlabel('Quarter')
ax2.set_ylabel('Unemployment (%)')
```

```python
# Just as a reference, here is what you get if you only define the first axes
# Still works fine, just does not show the second axes given we have not defined it

fig = plt.figure()
fig.suptitle('Unemployment Rates')

# Add an axes. It's the first axes of a 1 x 2 grid of axes.
ax1 = fig.add_subplot(1, 2, 1) # Number of rows, Number of columns, Position of this axes

# No need for a label this time
ax1.plot(data['NSW'].keys(), data['NSW'].values())

# Specify a title for the axes, and labels for the x- and y-axis
ax1.set_title('NSW')
ax1.set_xlabel('Quarter')
ax1.set_ylabel('Unemployment (%)')
```

### Enhancements

* Define Size
* Trim the range shown for the y-axis
* Define the tick range for the y-axis
* Show gridlines
* Let the second axes share the same y-axis as the with the first axes
* Specify the line color
* Define the amount of space between the axes (tight_layout)

```python
# Set the size to be 10 inches wide by 8 inches tall
fig = plt.figure(figsize=[10, 8])

fig.suptitle('Unemployment Rates')
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(data['NSW'].keys(), data['NSW'].values())
ax1.set_title('NSW')
ax1.set_xlabel('Quarter')
ax1.set_ylabel('Unemployment (%)')

# Set the y-axis values to go from 3 to 4
ax1.set_ylim(3, 4)

# Set the y-axis ticks to be 3.0, 3.1, 3.2, ..., 4.0 
ax1.set_yticks([x/10 for x in range(30, 41)])

# Show gridlines on the axes
ax1.grid()

# Tell ax2 to share its y-axis with ax1
ax2 = fig.add_subplot(1, 2, 2, sharey=ax1)

# Specify the colour of the line
ax2.plot(data['VIC'].keys(), data['VIC'].values(), color='green')

ax2.set_title('VIC')
ax2.set_xlabel('Quarter')

# Show gridlines on the axes
ax2.grid()

# Create a bit more space around the axes, to stop overlap
fig.tight_layout()
```

# SOPHISTICATED EXAMPLE

```python
fig = plt.figure(figsize=(8, 8))
fig.suptitle('A figure with three axes')

ax = fig.add_subplot(2, 2, 1)
ax.bar([1,2,3,4], [1,3,1,4])
ax.set_title('An axes with bar plot')
ax.set_xlabel('x-axis, with major ticks')
ax.set_ylabel('y-axis, with major ticks')

ax = fig.add_subplot(2, 2, 2)
ax.scatter([1,2,3,4], [2,4,1,2])
ax.set_title('An axes with scatter plot and grid')
ax.set_xlabel('x-axis, with major ticks')
ax.set_ylabel('y-axis, with major ticks')
ax.set_xlim(0,5)
ax.set_ylim(0,5)
ax.grid()

ax = fig.add_subplot(2, 1, 2)
ax.plot([1,2,3,4], [1,3,2,4], label='Line 1')
ax.plot([1,2,3,4], [2,1,4,3], label='Line 2')
ax.set_title('An axes with line plot')
ax.set_xlabel('x-axis, with major and minor ticks')
ax.set_ylabel('y-axis, with major and minor ticks')
ax.minorticks_on()
ax.legend(title='Legend')

fig.tight_layout()
```

Line plots are the default kind of plot for the plot method - they are what plot creates when you don't tell it otherwise.

```python
input_file = "C:\\Dropbox\\Variance\\UNSW\\ZZEN9021\\datafiles\weather.csv"
df = pd.read_csv(input_file)
s = df.groupby('month')['temp'].mean() # Find the mean temperature each month
```

## Multiple Axes

You can tell plot to put each column on its own axes, by adding subplots=True

```python
df.plot(
    kind = 'line',
    subplots = True,
)
```

Change it to a 1 x 3 layout

```python
df.plot(
    kind = 'line',
    subplots = True,
    layout = [1, 3], # 1 row of 3 columns
)
```

Notice that each axes has a different range of values on the y-axis. That's not helpful if we want to compare values across the axes. We can tell plot to get them to use the same range:

```python
df.plot(
    kind = 'line',
    subplots = True,
    layout = [1, 3],
    sharey = True, # Use the same y-axis
)
```

Similarly, you can tell plot to use the same range of values on each x-axis, by adding sharex = True.

## Seaborn

It's a bit of a pain having to map each species to a colour. Seaborn simplifies this process.

You can use the  relplot function of seaborn to create the scatter plot. It's not a method of a data frame - you have to call sns.replot, and you need to specify the data source using data=df. Also you need to set kind='scatter'.

```python
display(df.head())

sns.set() # Apply the default Seaborn style
sns.relplot(#Use a seaborn relplot
    data = df,
    kind = 'scatter',
    x = 'Sepal.Length',
    y = 'Petal.Length',
    hue = 'Species')
```

Notice that you can use the species column for the colour of the points (hue), even though the values in that column are not numerical - seaborn does the work for us.

Rather than distinguishing the species using colour, seaborn can distinguish them by size (size='Species'), or by marker (style='Species'). It can even put each species on a separate plot, either horizontally (col='Species') or vertically (row='Species').

```python
sns.set()
sns.relplot(
    data = df,
    kind = 'scatter',
    x = 'Sepal.Length',
    y = 'Petal.Length',
    hue = 'Species', # the color
    style = 'Species', # the type of marker
    row = 'Species' # Puts each species on a seperate subplot, in this case vertically as we said row (vs col)
            )
```

# Scatter Matrices

Scatter plots are useful for identifying a correlation between two data series.  The scatter_matrix function, part of the pandas.plotting library, creates a matrix of scatter plots comparing every possible pair of numerical columns of a data frame.  This simplifies the process of identifying potential correlations between different columns of a data frame.

```python
df[df['Species'] == 'virginica'].head()
```

So first it builds a matrix of row vs column.
Then it builds scatter plots for each combo.
Where the combo is of the same item (EG Sepal.Length vs Sepal.Length) it instead shows a histogram.

```python
pd.plotting.scatter_matrix (# Note: NOT a method of data frames
    df[df['Species'] == 'virginica'])

pd.plotting.scatter_matrix(
    df[df['Species'] ==  'setosa'])

pd.plotting.scatter_matrix(
    df[df['Species'] == 'versicolor'])
```

# Hexbin Plots

 A hexbin plot is a variation on a scatter plot.
    
Hexbin plots are good when you have a lot of overlapping data point.

The points are grouped into hexagonal bins, and the number of points within that bin are represented by a color - with ligher colors representing smaller numbers and darker colors representing larger numbers.

You can use the gridsize parameter to specify how many bins you would like along the x-axis (the number of bins along the y-axis is automatically adjusted).


```python
df.plot(
    kind = 'hexbin',
    x = 'Petal.Width',
    y = 'Sepal.Width',
    gridsize = 20, # specify the number of bins along the x-axis
    figsize = (10,6) # width x height in inches
    )
```
